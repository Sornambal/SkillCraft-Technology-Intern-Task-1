```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample data
data = {
    'Age Group': ['0 - 20 Years', '21 - 64 Years', '65+ Years'],
    'Population (Millions)': [512, 807, 98],
    'Percentage': [36.1, 57.0, 6.9]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Set style
sns.set(style="whitegrid")

# Plot bar chart
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x='Age Group', y='Population (Millions)', data=df, palette=['#f9e300', '#3cb5fc', '#ff5db0'])

# Annotate bars
for index, row in df.iterrows():
    plt.text(index, row['Population (Millions)'] + 10, f"{row['Percentage']}%",
             ha='center', va='bottom', fontweight='bold', fontsize=12)

# Title and labels
plt.title("India's Population Distribution by Age (2022)", fontsize=16, fontweight='bold')
plt.xlabel("Age Group", fontsize=12)
plt.ylabel("Population (in Millions)", fontsize=12)
plt.tight_layout()

# Show plot
plt.show()
```
